### Vanilla Gradient Descent

```python
y = np.loadtxt('Brachistochrone_data/y_all_plain.txt')
times = np.loadtxt('Brachistochrone_data/times_all_plain.txt')
```

```python
x = np.linspace(0, 10, y.shape[1], endpoint=True)
```

```python
plt.plot(np.linspace(1,len(times), len(times)), times, 'o--', color='blue')
plt.xlabel('Iterations')
plt.ylabel('Time')
plt.show()
```

```python
for i in range(len(y)):
    if i % 50 == 0:
        plt.figure(figsize=(8, 8))
        plt.plot(x, y[i], 'o--', color='red', label=f'Iteration #{i}')
        plt.plot(x, y[0], 'o--', label='Initial Guess', color='blue')
        plt.legend()
        plt.xlabel('x')
        plt.ylabel('y')
        plt.savefig(f'Brachistochrone_data/Brachistochrone_Plain/fig_{i}.png', dpi=300)
```
